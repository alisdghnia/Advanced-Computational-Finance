````python
import numpy as np
import scipy
````

````python
# your code

def histplot(df, ax, conf={'hist_color': 'lightblue', 'kde_color': 'crimson', 'bins': 10, 'hist_label': 'Histogram', 'kde_label': 'Normal PDF'}):
    sns.histplot(data=df, x='Consumer Discretionary', kde=False, stat='density', bins=conf['bins'], ax=ax, color=conf['hist_color'], label=conf['hist_label'])

    sns.kdeplot(data=df, x='Consumer Discretionary', color=conf['kde_color'], ax=ax, label=conf['kde_label'])
    # ax.set(xlim=(min(df['Consumer Discretionary']) - 0.001, max(df['Consumer Discretionary']) + 0.001))
    ax.legend()

fig, ax = plt.subplots(figsize=(7, 6))
histplot(df, ax)
````

````python
# your code
consumer_discretionary_data = df[['Consumer Discretionary']]

def histplot(df, ax, conf={'hist_color': 'lightblue', 'kde_color': 'crimson', 'bins': 10, 'hist_label': 'Histogram', 'kde_label': 'Normal PDF', 'ax2': True, 'stat': 'density'}):
    histogram = sns.histplot(df, x=df.columns[0], kde=False, stat=conf['stat'], bins=conf['bins'], ax=ax, color=conf['hist_color'], label=conf['hist_label'])
    if conf['ax2'] == True:
        ax2 = ax.twinx()
    else:
        ax2 = ax
    kde = sns.kdeplot(df, x=df.columns[0], color=conf['kde_color'], ax=ax2, label=conf['kde_label'])
    # ax.set(xlim=(min(df['Consumer Discretionary']) - 0.001, max(df['Consumer Discretionary']) + 0.001))
    if conf['ax2'] == True:
        handles, labels = histogram.get_legend_handles_labels()
        handles += kde.get_legend_handles_labels()[0]
        labels += kde.get_legend_handles_labels()[1]

        ax.legend(handles, labels, loc='upper right')
        ax2.set_ylabel('Probability Density')

        ax.grid(False)
        ax2.grid(False)

    else:
        ax.legend()

    ax2.set_xlim(float(df.min()), float(df.max()))

fig, ax = plt.subplots(figsize=(7, 6))
histplot(consumer_discretionary_data, ax)
````

````python
def histplot(df, ax, conf={'hist_color': 'lightblue', 'kde_color': 'crimson', 'bins': 10, 'hist_label': 'Histogram', 'kde_label': 'Normal PDF'}):
    sns.histplot(data=df, x='Consumer Discretionary', kde=False, stat='density', bins=conf['bins'], ax=ax, color=conf['hist_color'], label=conf['hist_label'])

    sns.kdeplot(data=df, x='Consumer Discretionary', color=conf['kde_color'], ax=ax, label=conf['kde_label'])
    # ax.set(xlim=(min(df['Consumer Discretionary']) - 0.001, max(df['Consumer Discretionary']) + 0.001))
    ax.legend()
    ax.grid(False)
    plt.xlabel('Weights')

fig, ax = plt.subplots(figsize=(7, 6))
histplot(df[df['Energy'] <= df['Energy'].mean()], ax, conf={'hist_color': 'orange', 'kde_color': 'blue', 'bins': 4, 'hist_label': 'Below Avg Energy Weights', 'kde_label': 'Normal PDF (Below Avg Energy Weights)'})
histplot(df[df['Energy'] > df['Energy'].mean()], ax, conf={'hist_color': 'lightblue', 'kde_color': 'crimson', 'bins': 4, 'hist_label': 'Above Avg Energy Weights', 'kde_label': 'Normal PDF (Above Avg Energy Weights)'})

ax.set_xlim(0.04, 0.135)
````
